# Importing Libraries

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Function for Image

```python
def load_image(image_path):
    image = cv2.imread(image_path)
    return image
```

# Function of Training

```python
def train_dataset(train_data, batches = 1000):
    for batch_start in range(0, len(train_data), batches):
        print(f"Processing batch {batch_start // batches + 1}")

        # Get the current batch
        batch_images = train_data.loc[batch_start:batch_start + batches - 1, "Path"]

        for image_name in batch_images:

            try:
                # Open image
                image = load_image(image_name)

            except Exception as e:
                print(f"Error processing {image_name}: {e}")
```
